# Solving for phase-matching ($\chi^{(3)}$)

* $\vec{k}_i = \frac{n_i \omega_i}{c} \hat{k}_i$ for i = 1,2,3.  The hat denotes a vector of length 1 i.e. each vector has a length of $n \omega$.
* Nonlinear polarization is created with spatial period of $\vec{k}_4 = \pm \vec{k}_1 \pm \vec{k}_2 \pm \vec{k}_3$ and a temporal period of $\omega_4=\pm \omega_1 \pm \omega_2 \pm \omega_3$.
* Light of frequency $\omega_4$ should propagate with spatial period of $|k_4| = n_4 \omega_4 / c$, but there $\vec{k}_4$ length is determined by the input wavevectors. 
* __phase mismatch__:  $\Delta k \equiv |k_4| - n_4 \omega_4 / c$ 

## DOVE Experiment

$$ \omega_4 = \omega_1 -\omega_2 + \omega_3 $$
$$ \vec{k}_4 = \vec{k}_1 - \vec{k}_2 + \vec{k}_3 $$

We will consider $\omega_4 > \omega_3 > \omega_2 > \omega_1$.

## __BOXCARS__ geometry:

\begin{align*}
\vec{k}_1 &= n_1 \omega_1 \begin{bmatrix}
        \sin \theta_1 \\
        0 \\
        \cos \theta_1 \\
        \end{bmatrix}
\end{align*}

\begin{align*}
\vec{k}_2 &= n_2 \omega_2 \begin{bmatrix}
        0 \\
        \sin \theta_2 \\
        \cos \theta_2 \\
        \end{bmatrix}
\end{align*}


\begin{align*}
\vec{k}_3 &= n_3 \omega_3 \begin{bmatrix}
        \sin \theta_3 \\
        0 \\
        \cos \theta_3 \\
        \end{bmatrix}
\end{align*}


For simpler notation, define $c_i \equiv \frac{n_i \omega_i}{c} \cos \theta_i$ and $s_i \equiv \frac{n_i \omega_i}{c} \sin \theta_i$.  Then

\begin{align*}
\vec{k}_4 &= \begin{bmatrix}
        s_1 + s_3 \\
        -s_2 \\
        c_1 - c_2 + c_3 \\
        \end{bmatrix}
\end{align*}



The BOXCARS geometry specifies that the plane defined by $k_1$ and $k_3$ are normal to the plane defined by $k_2$ and $k_4$.  In mathematical terms, we can write this constraint as $(k_1 \times k_3) \cdot (k_2 \times k_4) = 0$, where $\times$ represents the cross product.  

\begin{align*}
k_1 \times k_3 &= \begin{bmatrix}
        0 \\
        c_1 s_3 - c_3 s_1 \\
        0 \\
        \end{bmatrix}
\end{align*}

\begin{align*}
k_2 \times k_4 &= \begin{bmatrix}
        -s_2 c_2 - (c_1 - c_2 + c_3)s_2 \\
        -(s_1 + s_3)c_2 \\
        (s_1 + s_3) s_2 \\
        \end{bmatrix}
\end{align*}

Assuming that $k_1$ and $k_3$ are not collinear (i.e. $c_1 s_3 - s_1 c_3 = 0$ ), $k_1$ and $k_3$ lie in the $xz$ plane, so their cross product points in the $y$ direction.
This means $k_2 \times k_4$ must lie in the $xz$ plane i.e.

$$ (s_1 + s_3) c_2 = 0 $$

One solution to this is $c_2 = 0$, i.e. $k_2$ runs along the $y$ axis.  This is a huge crossing angle and we will not consider it here.
Instead, __we consider the constraint__ $s_1 = -s_3$.  Note that this also fixes the relation between $c_1$ and $c_3$:  TODO


With this definition, it is easy to verify that the $x$ component of $k_4$ goes to zero, i.e.

\begin{align*}
\vec{k}_4 &= \begin{bmatrix}
        0 \\
        -s_2 \\
        c_1 - c_2 + c_3 \\
        \end{bmatrix}
\end{align*}

Importantly, note that $k_4$ does not generally fixed in a specific plane; it has $x$, $y$, and $z$ components.
Phase matching solutions for the more general defintion of $k_4$ (see above) is not tied to this plane.  
In general, if the phase-matching is aligned in a BOXCARS geometry and then input colors are changed without altering the direction of _all three_ input beams, the new phase matching direction will not generally lie in the $yz$ plane.


### Solving the phase matching condition

From the above, we have

\begin{align*}
    \left| \vec{k}_4 \right|^2 
        &= s_2^2 + (c_1 - c_2 + c_3)^2  \\
        &= \frac{n_2^2\omega_2^2}{c^2} - c_2^2 + (c_1 - c_2 + c_3)^2
\end{align*}

Setting $|k_4| = n_4 \omega_4 / c$ (i.e. phase-matching), we have

\begin{align*}
    \frac{n_4^2}{c^2} (\omega_1 - \omega_2 + \omega_3)^2 - \frac{n_2^2\omega_2^2}{c^2}
        &= (c_1 - c_2 + c_3)^2 - c_2^2
\end{align*}

For convenience, __we will now work in units such that__ $c=1$.

$$
s_1 s_3 - c_1 c_2 + c_1 c_3 - c_2 c_3 = n_4^2 (\omega_1 \omega_3 - \omega_1 \omega_2 - \omega_2 \omega_3) + \frac{n_4^2 - n_1^2}{2} \omega_1^2 + \frac{n_4^2 - n_2^2}{2} \omega_2^2 + \frac{n_4^2 - n_3^2}{2} \omega_3^2
$$


Noting that $s_1 s_3 = -s_3^2 = c_3^2 - n_3^2 \omega_3^2$, we have

$$
c_1^2 - c_1 c_2 + c_1 c_3 - c_2 c_3 = n_4^2 (\omega_1 \omega_3 - \omega_1 \omega_2 - \omega_2 \omega_3) + \frac{n_4^2 + n_1^2}{2} \omega_1^2 + \frac{n_4^2 - n_2^2}{2} \omega_2^2 + \frac{n_4^2 - n_3^2}{2} \omega_3^2
$$



Defining the LHS as $A$, we can express the phase matching equation as a quadratic-like equation:
$$ c_3^2 + (c_1 - c_2) c_3 - (c_1 c_2 + A) = 0 $$
Note that in this equation, $c_1$ is a function of $c_3$: $c_1 = \sqrt{n_1^2 \omega_1^2 - n_3^2 \omega_3^2 + c_3^2}$ by our BOXCARS geometry, so the equation is not strictly quadratic in $c_3$ and we cannot find the roots analytically. 

There are two free variables in this equation: $c_3$ and $c_2$.  We can set $c_2$ and use a [root-finding program](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fsolve.html) to determine the solution.


### Further directions:  fix output direction

If we desire a fixed output direction, we can put constraints on the output vector e.g. define $\theta_4$ such that

\begin{align*}
\vec{k}_4 &= n_4 \omega_4 \begin{bmatrix}
        0 \\
        \sin\theta_4 \\
        \cos\theta_4 \\
        \end{bmatrix}
\end{align*}

Then

\begin{align*}
    \tan\theta_4 &= \frac{-s_2}{c_1 - c_2 + c_3} \\
    \implies \left| k_4 \right|^2 &= ( 1+\tan^2\theta_4 ) (c_1 - c_2 + c_3)^2
    \implies (c_1 - c_2 + c_3)^2 &= \frac{n_4^2 (\omega_1 - \omega_2 + \omega_3)^2}{1 + \tan^2\theta_4}
\end{align*}

Note, however, that _such a constraint would require changing all three input angles_.  We will not apply this constraint in our following derivations.

## compute solution

```python
import numpy as np
from scipy import optimize
```

```python


def phase_matching_angles(ns:list, ws:list, theta2:float):
    n1, n2, n3, n4 = ns
    w1, w2, w3, w4 = ws
    assert round(w4, 5) == round(w1 - w2 + w3, 5)
    A = n4**2 * (w1*w3 - w1*w2 - w2*w3) + 0.5 * (
        (n4**2 + n1**2) * w1**2 + (n4**2 - n2**2) * w2**2 + (n4**2 - n3**2) * w3**2
    )
    c2 = n2 * w2 * np.cos(theta2)
    c3 = lambda c1: np.sqrt(n3**2 * w3**2 - n1**2 * w1**2 + c1**2)
    def pm(x):
        # coerce x to positive cosine values
        x = max([0, x])
        x = min([np.pi/2, x])
        c1 = n1 * w1 * np.cos(x)
        # print(c3(c1))
        return c1**2 + c1 * (c3(c1) - c2) - c2*c3(c1) - A

    theta1 = optimize.fsolve(pm, 5 * np.pi/180)[0]
    theta3 = np.arcsin(-n1 * w1 * np.sin(theta1) / (n3 * w3))

    return theta1, theta2, theta3


```

```python
def ks(ns, ws, thetas):
    n1, n2, n3, *_ = ns
    w1, w2, w3, *_ = ws
    t1, t2, t3, *_ = thetas
    k1 = n1 * w1 * np.array([np.sin(t1), 0, np.cos(t1)])
    k2 = n2 * w2 * np.array([0, np.sin(t2), np.cos(t2)])
    k3 = n3 * w3 * np.array([np.sin(t3), 0, np.cos(t3)])
    return k1, k2, k3, k1 - k2 + k3

def phase_mismatch(ns, ws, thetas):
    k4_length = np.linalg.norm(ks(ns, ws, thetas)[3])
    #  np.sqrt( (ss[0] + ss[2])**2 + ss[1]**2 + (cs[0] - cs[1] + cs[2])**2 )**0.5
    return k4_length - ns[3] * ws[3]
```

```python
wn_to_omega = lambda wn: 2*np.pi*wn  # remember, c=1 units
ws = [3000, 1500, 7000]
ws += [ws[0] - ws[1] + ws[2]]
print(ns, ws)
ns = [1.5] * 4

thetas = phase_matching_angles(ns, list(wi for wi in map(wn_to_omega, ws)), theta2=5 * np.pi/180)
print(thetas)
for ki in ks(ns, ws, thetas):
    print(ki)
mismatch = phase_mismatch(ns, ws, thetas)
print(mismatch)
```

```python

def grid_thetas(w1, w2, w3):
    thetas = np.zeros((w1.size, w2.size, w3.size, 4))

    for ndi in np.ndindex(thetas.shape[:3]):
        color_combination = [
            w_x[ndi_x] for (w_x, ndi_x) in zip([w1, w2, w3], ndi)
        ]
        color_combination += [color_combination[0] - color_combination[1] + color_combination[2]]
        ns = [1.5, 1.5, 1.5, 1.5]
        ws = list(wi for wi in map(wn_to_omega, color_combination))

        try:
            thetas[ndi][:3] = phase_matching_angles(ns, ws, theta2=5 * np.pi/180)
            k4 = ks(ns, ws, thetas[ndi][:3])[3]
            thetas[ndi][-1] = np.arctan2(k4[1], k4[2])
        except AssertionError:
            print(color_combination)
            return
        # print(ndi, color_combination, thetas[ndi])
    return thetas

```
